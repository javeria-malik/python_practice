```python
import numpy as np
import matplotlib.pyplot as plt

# Set the scale parameter for the Rayleigh distribution
sigma = 1.0

# Generate random numbers from a Rayleigh distribution
random_numbers = np.random.rayleigh(scale=sigma, size=1000)

# Plot a histogram of the generated random numbers
plt.hist(random_numbers, bins=30, density=True, alpha=0.7, label='Generated Data')

# Define x values for the PDF curve
x_values = np.linspace(0, 5, 1000)

# Calculate the Rayleigh PDF values for the defined x values
pdf_values = (x_values / sigma**2) * np.exp(-x_values**2 / (2 * sigma**2))

# Plot the Rayleigh PDF curve
plt.plot(x_values, pdf_values, label=f'Rayleigh PDF, sigma={sigma}', linewidth=2)

# Set plot labels and title
plt.title('Random Number Generation from Rayleigh Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import rayleigh
from scipy.optimize import minimize

# Generate a sample dataset
np.random.seed(42)
sample_data = np.random.rayleigh(scale=2, size=1000)  # Fixed the scale and size parameters

# Define the negative log-likelihood function for Rayleigh distribution
def neg_log_likelihood(params, data):
    sigma = params[0]
    return -np.sum(rayleigh.logpdf(data, scale=sigma))

# Initial guess for the parameter
initial_guess = [1.0]  # Added the missing '=' sign

# Minimize the negative log-likelihood function to estimate the parameter
result = minimize(neg_log_likelihood, initial_guess, args=(sample_data,))  
estimated_sigma = result.x[0]  # Adjusted the index to get the estimated sigma

# Plot the histogram of the data and the fitted Rayleigh distribution
x_values = np.linspace(0, max(sample_data), 1000)  # Added x_values for plotting
plt.hist(sample_data, bins=30, density=True, alpha=0.7, label='Sample Data')
plt.plot(x_values, rayleigh.pdf(x_values, scale=estimated_sigma), label='Fitted Rayleigh PDF, sigma={:.3f}'.format(estimated_sigma), linewidth=2)
plt.title('Parameter Estimation for Rayleigh Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```
